```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import jieba
import cv2

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#text = "人工智能 人工智能 人工智能 人工智能 人工智能 人工智能 人工智能 人工智能 人工智能 预测 预测 预测 预测 预测 预测 预测 预测 运筹 运筹 运筹 运筹 运筹 \
#优化 优化 优化 优化 优化 优化 优化 优化 优化 效率 效率 效率 效率 效率 效率 效率 动态调价 人员排班 动态调价 人员排班"

text = open("ppttxt.txt").read()
text = " ".join(jieba.cut(text))
```

```python

# 读取蒙板/彩色图像
#logo_coloring = np.array(Image.open("logo_no_back.png"))
#stopwords = set(STOPWORDS)
#stopwords.add("said")

logo_coloring = cv2.imread('logo_no_back.png')

wc = WordCloud(font_path="xuyuan.ttf",
               background_color="white", max_words=200, 
               mask=logo_coloring,
               #stopwords=stopwords, 
               max_font_size=80, 
               #random_state=42
              )
# 生成词云
wc.generate(text)

# 从图像创建着色
image_colors = ImageColorGenerator(logo_coloring)
```

```python
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# 显示
plt.imshow(wc, interpolation="bilinear")
plt.axis("off") #不显示坐标尺寸
plt.figure()
# 重新着色词云并显示
# 我们也可以直接在构造函数中给使用：color_func=image_colors 
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off") #不显示坐标尺寸
plt.figure()
plt.imshow(logo_coloring, cmap=plt.cm.gray, interpolation="bilinear")
plt.axis("off") #不显示坐标尺寸
plt.show()#一次绘制三张图
```
